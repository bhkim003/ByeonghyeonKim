```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv("bh_chan_clustering_train_main fusion5 chan_loss_factor sweep hidden2.csv")

# 필요한 열 추출
x = df["chan_loss_factor"]  # x축: spike_aware_factor
y = df["k_means_acc_best"]  # y축: accuracy

# 그래프 그리기
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Accuracy')
plt.xlabel("Spike Aware Factor")
plt.ylabel("Accuracy")
plt.ylim(80, 95)  # y축 범위 설정
plt.title("Spike Aware Factor vs Accuracy, Epoch-20, hidden-2")
plt.grid(True)
plt.legend()
plt.show()
```
