```python
import os
import re

# 분석 대상 키워드
keywords = [
    "lif layer 1 self.abs_max_v:",
    "lif layer 2 self.abs_max_v:",
    "fc layer 1 self.abs_max_out:",
    "fc layer 2 self.abs_max_out:",
    "fc layer 3 self.abs_max_out:",
    "max_activation_accul updated:",
]

# 로그별 결과 저장
results = {}

# 전체 최댓값 저장
global_max = {key: float("-inf") for key in keywords}

# 파일별 처리
for i in range(10):
    file_path = f"output{i}.log"
    if not os.path.exists(file_path):
        print(f"⚠️ 파일 없음: {file_path}")
        continue

    # 키워드별 최댓값 초기화
    max_values = {key: float("-inf") for key in keywords}

    with open(file_path, "r") as f:
        for line in f:
            for key in keywords:
                if key in line:
                    match = re.search(rf"{re.escape(key)}\s*([-+]?\d*\.?\d+(?:[eE][-+]?\d+)?)", line)
                    if match:
                        value = float(match.group(1))
                        if value > max_values[key]:
                            max_values[key] = value
                        if value > global_max[key]:
                            global_max[key] = value

    # 결과 저장
    results[f"output{i}.log"] = max_values

# 로그별 결과 출력
for logfile, max_values in results.items():
    print(f"\n=== {logfile} ===")
    for key, val in max_values.items():
        if val == float("-inf"):
            print(f"{key} 데이터 없음")
        else:
            print(f"{key} {val}")

# 전체 결과 출력
print("\n=== 전체 로그에서의 최댓값 ===")
for key, val in global_max.items():
    if val == float("-inf"):
        print(f"{key} 데이터 없음")
    else:
        print(f"{key} {val}")
```
