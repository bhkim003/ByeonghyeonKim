```python
import csv
import numpy as np

def load_csv_column(file_path):
    """CSV 파일에서 두 번째 열을 NumPy 배열로 로드"""
    values = []
    with open(file_path, newline="", encoding="utf-8") as f:
        reader = csv.reader(f)
        next(reader)  # 헤더 건너뛰기
        for row in reader:
            values.append(float(row[1]))  # 두 번째 열 가져오기
    return np.array(values)

# 기존 데이터 로드
acc = load_csv_column("ann_01norm_50quant_acc.csv")
loss = load_csv_column("ann_01norm_50quant_loss.csv")

# postpretrain 데이터 로드
acc_postpretrain = load_csv_column("ann_01norm_50quant_acc_postpretrain.csv")
loss_postpretrain = load_csv_column("ann_01norm_50quant_loss_postpretrain.csv")

# 데이터 합치기
acc = np.concatenate([acc, acc_postpretrain])
loss = np.concatenate([loss, loss_postpretrain])

print(acc.shape, loss.shape)




import matplotlib.pyplot as plt

# 산점도 그리기
plt.scatter(loss, acc, alpha=0.5)

# 축 레이블 설정
plt.xlabel("Loss")
plt.ylabel("Accuracy")

# 제목 설정
plt.title("Scatter Plot of Accuracy vs. Loss")

# 그래프 표시
plt.show()



import numpy as np
import matplotlib.pyplot as plt

loss_cut = 0.012
mask = loss < loss_cut
filtered_loss = loss[mask]
filtered_acc = acc[mask]

# 산점도 그리기
plt.scatter(filtered_loss, filtered_acc, alpha=0.5)

# x축을 로그 스케일로 설정
plt.xscale("log")

# 축 레이블 설정
plt.xlabel("Loss (log scale)")
plt.ylabel("Accuracy")

# 제목 설정
plt.title(f"Scatter Plot of Accuracy vs. Loss (Log Scale, Loss < {loss_cut})")

# 그래프 표시
plt.show()

```
