```python
import csv
import numpy as np

second_column_values = []

# CSV 파일 열기
with open("ann_01norm_50quant_acc.csv", newline="", encoding="utf-8") as f:
# with open("ann_01norm_50quant_acc_postpretrain.csv", newline="", encoding="utf-8") as f:
    reader = csv.reader(f)
    next(reader)  # 헤더 건너뛰기
    
    for row in reader:
        second_column_values.append(float(row[1]))  # 두 번째 열 가져오기

# NumPy 배열로 변환
acc = np.array(second_column_values)

print(acc)  # shape: (N,)

print(acc.shape)

import csv
import numpy as np

second_column_values = []

# CSV 파일 열기
with open("ann_01norm_50quant_loss.csv", newline="", encoding="utf-8") as f:
# with open("ann_01norm_50quant_loss_postpretrain.csv", newline="", encoding="utf-8") as f:
    reader = csv.reader(f)
    next(reader)  # 헤더 건너뛰기
    
    for row in reader:
        second_column_values.append(float(row[1]))  # 두 번째 열 가져오기

# NumPy 배열로 변환
loss = np.array(second_column_values)

print(loss)  # shape: (N,)

print(loss.shape)



import matplotlib.pyplot as plt

# 산점도 그리기
plt.scatter(loss, acc, alpha=0.5)

# 축 레이블 설정
plt.xlabel("Loss")
plt.ylabel("Accuracy")

# 제목 설정
plt.title("Scatter Plot of Accuracy vs. Loss")

# 그래프 표시
plt.show()



import numpy as np
import matplotlib.pyplot as plt

loss_cut = 0.012
mask = loss < loss_cut
filtered_loss = loss[mask]
filtered_acc = acc[mask]

# 산점도 그리기
plt.scatter(filtered_loss, filtered_acc, alpha=0.5)

# x축을 로그 스케일로 설정
plt.xscale("log")

# 축 레이블 설정
plt.xlabel("Loss (log scale)")
plt.ylabel("Accuracy")

# 제목 설정
plt.title(f"Scatter Plot of Accuracy vs. Loss (Log Scale, Loss < {loss_cut})")

# 그래프 표시
plt.show()

```
